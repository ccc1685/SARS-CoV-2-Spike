```julia
S1 = readdlm("S1.txt")[:,1];
S2 = readdlm("S2.txt")[:,1];
```

```julia
alphasubs = [(501,"Y"),(570,"D"),(614,"G"),(681,"H"),(716,"I"),(982,"A"),(1118,"H")]
alphadeletes = [(144,144),(69,70)]
betasubs = [(18,"F"),(80,"A"),(215,"G"),(246,"I"),(417,"N"),(484,"K"),(501,"Y"),(614,"G"),(701,"V")]
betadeletes = [(242,244)]
gammasubs = [(18,"F"),(20,"N"),(26,"S"),(138,"Y"),(190,"S"),(417,"T"),(484,"K"),(501,"Y"),(614,"G"),(655,"Y"),(1027,"I"),(1176,"F")]
gammadeletes =[]
deltasubs =  [(19,"R"),(452,"R"),(478,"K"),(614,"G"),(681,"R"),(950,"N")]
deltadeletes =[(157,158)]
```

```julia
Salpha = mutation(S2,alphasubs,alphadeletes);
Sbeta = mutation(S2,betasubs,betadeletes);
Sgamma = mutation(S2,gammasubs,gammadeletes);
Sdelta = mutation(S2,deltasubs,deltadeletes);
```

```julia
range,sds,m=scan_window(S1)
w1 = plot(range,sds,xlabel="W",ylabel="Hydropathy",lw=2,label="CoV-1")
range,sds,m=scan_window(S2)
w2 = plot!(range,sds,xlabel="W",ylabel="Hydropathy",lw=2,label="CoV-2")
range,sds,m=scan_window(Salpha)
walpha = plot!(range,sds,xlabel="W",ylabel="Hydropathy",lw=2,label="Alpha")
range,sds,m=scan_window(Sdelta)
wdelta = plot!(range,sds,xlabel="W",ylabel="Hydropathy",lw=2,label="Delta")
savefig("W")
```

```julia
S = Sbeta
range,sds,m=scan_window(S)
psi = score(S,MZ,m[2]+10);
out=findminima(psi,40)
```

```julia
S = Salpha
tle = "Alpha"
range,sds,m=scan_window(S)
psi = score(S,MZ,m[2]+10);
out=findminima(psi,40)
palpha = plot(psi[:,1],psi[:,2],xlabel="residue",ylabel="Hydropathy",title=tle,legend=false,lw=2,color=:black,ylims=(130,200))
```

```julia
S = Sdelta
tle = "Delta"
range,sds,m=scan_window(S)
psi = score(S,MZ,m[2]+10);
out=findminima(psi,40)
pdelta = plot(psi[:,1],psi[:,2],xlabel="residue",title=tle,legend=false,lw=2,color=:black,ylims=(130,200))
```

```julia
S = S1
tle = "Cov-1"
range,sds,m=scan_window(S)
psi = score(S,MZ,m[2]+10);
out=findminima(psi,40)
p1 = plot(psi[:,1],psi[:,2],ylabel="Hydropathy",title=tle,legend=false,lw=2,color=:black,ylims=(130,200))
```

```julia
S = S2
tle = "Cov-2"
range,sds,m=scan_window(S)
psi = score(S,MZ,m[2]+10);
out=findminima(psi,40)
p2 = plot(psi[:,1],psi[:,2],title=tle,legend=false,lw=2,color=:black,ylims=(130,200))
```

```julia
plot(p1, p2, palpha, pdelta, layout = (2, 2), legend = false)
# savefig("H.pdf")
```
